```
top100_df = pd.DataFrame(top100, columns=['strDrink'])
top100_df.index = top100_df.index.set_names(['rank'])
top100_df = top100_df.reset_index()
top100_df
```

```
filtered_df = df[df["strDrink"].isin(top100)]
filtered_df
```

```
top_df = filtered_df[['strDrink', 'strCategory', 'strGlass', 'strIngredients', 'strInstructions','Alc_type']].reset_index(drop=True)
top_df
```

```
merge_df = pd.merge(top_df, top100_df, on='strDrink', how='left')
merge_df
```

```
merge_df.columns
```

```
group_df = merge_df.groupby('strDrink')['strIngredients', 'Alc_type','rank','strInstructions'].agg({
    'strIngredients': lambda x: list(x.dropna().unique()),
    'Alc_type': lambda x: list(x.dropna().unique()),
    'rank': 'first',
    'strInstructions': 'first'
}).reset_index()

group_df
```

```
group_df.to_json('grouped_cocktails.json', orient='records', indent=2)
```

```
# Danni
unique_ingredients = set()

for ingredients_list in group_df['strIngredients']:
    unique_ingredients.update(ingredients_list)

print(len(unique_ingredients))
```
