```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv("data_cocktails_cleaned.csv")
df
```

```
group_df = df.groupby('strDrink')['strGlass','strCategory', 'strIngredients', 'Alc_type', 'Basic_taste','strInstructions'].agg({
    'strGlass':"first",
    'strCategory':"first",
    'strIngredients': lambda x: list(x.dropna().unique()),
    'Alc_type': lambda x: list(x.dropna().unique()),
    'Basic_taste': lambda x: list(x.dropna().unique()),
    'strInstructions': "first"
}).reset_index()

group_df
```

```
df = pd.read_csv("data_cocktails_cleaned.csv")
df
```

```
df["Garnish_type"].value_counts()
```

```
df1 = df[~df["Garnish_type"].isna()]
df1
```

```
df2 = df1[df1["Garnish_type"] != "top up"]
df2
```

```
df2.columns
```

```
df2['strIngredients'].value_counts().to_dict()
```

```
df2[df2['strGlass'] == "Whiskey sour glass"]["strIngredients"].value_counts().to_dict()
```

```
group_df = df2.groupby('strGlass').agg({
    'strGlass': 'size',
    'strDrink': lambda x: list(x.dropna().unique()),
    'strIngredients': lambda x: list(x.dropna().unique())    
}).rename(columns={'strGlass': 'number_glass'}).reset_index()

group_df
```

```
df_clean = group_df[group_df["number_glass"] > 2].reset_index(drop = True)
df_clean
```

```
df_clean["strGlass"]
```

```
df_clean["strDrink"][6]
```

```
df_clean.to_json('treemap_data.json', orient='records', indent=2)
```
