```
group_df = df.groupby('strDrink')['strGlass','strCategory', 'strIngredients', 'Alc_type', 'Basic_taste','strInstructions'].agg({
    'strGlass':"first",
    'strCategory':"first",
    'strIngredients': lambda x: list(x.dropna().unique()),
    'Alc_type': lambda x: list(x.dropna().unique()),
    'Basic_taste': lambda x: list(x.dropna().unique()),
    'strInstructions': "first"
}).reset_index()

group_df
```

```
group_df[group_df['Basic_taste'].apply(lambda x: isinstance(x, list) and len(x) == 0)]
```

```
filter_df = group_df[group_df['Basic_taste'].apply(lambda x: isinstance(x, list) and len(x) != 0)]
filter_df
```

```
filter_df[filter_df['Alc_type'].apply(lambda x: isinstance(x, list) and len(x) == 0)]
```

```
all_drink_clean = filter_df[filter_df['Alc_type'].apply(lambda x: isinstance(x, list) and len(x) != 0)]
all_drink_clean
```

```
all_drink_clean.info()
```

```
len(all_drink_clean["strGlass"].unique())
```

```
all_drink_clean.to_json('all_drink_clean.json', orient='records', indent=2)
```
